```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
X=kidney_df.drop(columns='Class',axis=1)
y=kidney_df['Class']
```

```python
print(X)
```

```python
print(y)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,stratify=y,random_state=42)
```

```python
X.shape,X_train.shape,X_test.shape
```

```python
from sklearn import svm
classifier = svm.SVC(kernel='linear')
classifier.fit(X_train, y_train)
```

```python
X_train_pred=classifier.predict(X_train)
training_data_accuracy=accuracy_score(X_train_pred,y_train)
print("Accuracy Score of Train data:",training_data_accuracy)
```

```python
X_test_pred=classifier.predict(X_test)
testing_data_accuracy=accuracy_score(X_test_pred,y_test)
print("Accuracy Score of Test data:",testing_data_accuracy)
```

```python
input_data=(70.0,1.02,0.0,0.0,1.0,26.0,1.1,137.53,4.63,17.0,6700.0,6.1,0.0)

#creating array
input_data_as_np_array=np.asarray(input_data)

#reshape
input_data_reshaped=input_data_as_np_array.reshape(1,-1)

#prediction
prediction=classifier.predict(input_data_reshaped)

if (prediction[0]==0):
    print("Person  not have a Kidney disease ")
else:
    print("Person has Kidney disease")
```
