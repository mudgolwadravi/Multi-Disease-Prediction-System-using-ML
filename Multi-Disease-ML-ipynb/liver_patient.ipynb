```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
liver_df=pd.read_csv('C:/Users/mudgo/OneDrive/Desktop/ml prj/datasets/indian_liver_patient.csv')
liver_df.tail()
```

```python
liver_df.dropna(inplace=True)
```

```python
liver_df.isnull().sum()
```

```python
liver_df['Gender']=liver_df['Gender'].map({'Male':1,'Female':0})
```

```python
X=liver_df.drop(columns='Dataset',axis=1)
y=liver_df['Dataset']
```

```python
print(X)
```

```python
print(y)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,stratify=y,random_state=42)
```

```python
X.shape,X_train.shape,X_test.shape
```

```python
# from sklearn import svm
# classifier = svm.SVC(kernel='linear')
# classifier.fit(X_train, y_train)

from sklearn.ensemble import RandomForestClassifier
classifier=RandomForestClassifier(n_estimators=100,max_depth=6)
classifier.fit(X_train,y_train)
```

```python
X_train_pred=classifier.predict(X_train)
training_data_accuracy=accuracy_score(X_train_pred,y_train)
print("Accuracy Score of Train data:",training_data_accuracy)
```

```python
X_test_pred=classifier.predict(X_test)
testing_data_accuracy=accuracy_score(X_test_pred,y_test)
print("Accuracy Score of Test data:",testing_data_accuracy)
```

```python
input_data=(30,1,1.3,0.4,482,102,80,6.9,3.3,0.9)

#creating array
input_data_as_np_array=np.asarray(input_data)

#reshape
input_data_reshaped=input_data_as_np_array.reshape(1,-1)

#prediction
prediction=classifier.predict(input_data_reshaped)

if (prediction[0]==2):
    print("Person  not  a Liver patient ")
else:
    print("Person is Liver patient  ")
```
